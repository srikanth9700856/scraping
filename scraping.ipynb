```python
from bs4 import BeautifulSoup
import urllib    # it is used for opening and reading urls

url ="https://www.idrbt.ac.in/ipdf.html"
```

```python
req = urllib.request.Request(url) #requesting url
data = urllib.request.urlopen(req).read()
```

```python
soup= BeautifulSoup(data, "html.parser")
```

```python
main_content= soup.find("div", attrs= {"id" : "page-main"})
```

```python
lists = main_content.find_all("p")
```

```python
str = ""
for list in lists:
    info= list.text
    str+=info
```

```python
#Nouns

import nltk
nltk.download('averaged_perceptron_tagger')
nltk.download('punkt')

lines  = str

# function to test if something is a noun
is_noun = lambda pos: pos[:2] == 'NN'

# do the nlp stuff
tokenized = nltk.word_tokenize(lines)

extracted_nouns = [word for (word, pos) in nltk.pos_tag(tokenized) if is_noun(pos)] 
nouns = ' '.join([elem for elem in extracted_nouns]) 

print(nouns)

```

```python
# WordCloud of Nouns
wordcloud = WordCloud(max_font_size=800, background_color="white").generate(nouns)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
